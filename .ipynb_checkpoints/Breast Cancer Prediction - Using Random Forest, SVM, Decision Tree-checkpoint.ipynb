```python
# in this tutorial we will be using machine leanring to predict the possibility of getting breast cancer using kaggle dataset
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # data visualization
import seaborn as sns #data visualization
```

```python
df = pd.read_csv("breast-cancer.csv") #reading the csv file
df.head() #showing the csv file
```

```python
df.isnull().sum() #checking if the columns is hvaing null values
```

```python
df.describe()
```

```python
# dropping unwanted columns which are not needed for Ml algorithm
df=df.drop("id",axis='columns')
df.head()
```

```python
# PLotting the scatterplot on diagnosis columns
sns.scatterplot(data = df, x= 'smoothness_mean', y='texture_mean', hue='diagnosis') 
```

```python
# if we observe the data the diagnosis columns is having the alphabetic values hence we are converting it to numberic as out algorith works better with numberic values
dia=pd.get_dummies(df.diagnosis,drop_first=True) # Converting the diagnosis columns to a numberic field which has value one if the value is M and value 0 if its B
dia
df1 = pd.concat([df,dia],axis='columns') # merging the dummies field with orginal data frame
df1=df1.drop(['diagnosis'],axis='columns') #dropping the gender columns from the data frame
df1.head()
```

```python
# creating two data set for training and testing the data
x=df1.drop("M",axis='columns')
y=df1["M"]
```

```python
# Calling library for Ml algorithm - Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
model = RandomForestClassifier(n_estimators=350)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test) # Score of using Random Forest
```

```python
# Calling library for Ml algorithm - SVM
from sklearn.svm import SVC 
clf=SVC()
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test) # Score of Using Single Vector Machine
```

```python
# Calling library for Ml algorithm - Decision Tree
from sklearn.tree import DecisionTreeClassifier 
dt = DecisionTreeClassifier(random_state=0)
dt.fit(X_train, y_train)
```

```python
dt.score(X_test,y_test) # Score of Using Decision Tree
```

# Here we can see Random Forest is yeilding better score compared to SVM and DT
